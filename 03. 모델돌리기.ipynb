```python
# k-fold
```

```python
from sklearn.model_selection import KFold
```

```python
k = KFold(n_splits=5,shuffle=True,random_state=777)
# 5개로 쪼개서 진행. 무작위로 섞어서
```

```python
for train_idx,val_idx in k.split(X_train):
    print(len(train_idx),len(val_idx))
```

```python
model  = lgb.LGBMRegressor(random_state=777,n_estimators=1000)
```

```python
models=[]
```

```python
for train_idx,val_idx in k.split(X_train):
    x_t = X_train.iloc[train_idx]
    y_t = y_train.iloc[train_idx]
    x_val = X_train.iloc[val_idx]
    y_val = y_train.iloc[val_idx]
    
    models.append(model.fit(x_t,y_t,eval_set=(x_val,y_val),early_stopping_rounds=100,verbose=100))
 
# earlystopping_rounds 
# x_t와 y_t를 학습시켜 eval_set과 비교해서 오차가 크게 나지 않으면 일찍 멈춘다. 
# verbose = 100번마다 한번 씩 모델에 대한 결과 프린트
```

```python
preds = []
for model in models:
    preds.append(model.predict(X_test))
    
```

```python
kk= submission.copy()
```

```python
preds
```

```python
kk['box_off_num'] = np.mean(preds,axis=0)
```

```python
kk
```

```python
kk.to_csv('result02.csv',index=False)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model  = lgb.LGBMRegressor(random_state=777,n_estimators=1000)
```

```python
params = {
    'learning_rate':[0.1,0.01,0.003],
    'min_child_samples':[20,30]}

gs = GridSearchCV(estimator = model,
                 param_grid=params,
                 scoring='neg_mean_squared_error',
                 cv=k)
# scoring = 'rmse'
```

```python
gs.fit(X_train,y_train)
```

```python
# 가장 성능이 좋았던 베스트 파라미터 출력
gs.best_params_
```

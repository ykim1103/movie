### 범주형 데이터 인코딩

##### 인코딩을 위한 concat!

```python
# train 데이터에 있는 마지막열 제외(관객수)
train_encoding = train.iloc[:,0:-1]
```

```python
train_test["new_num"]=0
for i in list(a.keys()):
    train_test.loc[train_test["distributor"] == i,"new_num"] = a[i]
train_test

####--------------------------------------------
# i에는 a에서 받은 키값들(배급사이름)이 들어온다. 
# 즉, i와 train_test의 배급사명이 일치하면 그 배급사에 
# 해당하는 value값이 new_num에 들어오게끔 했다.
# a는 dict이다. 
# a[키값] = 해당 키값에 해당하는 value값.
```

### new_num(배급사별 영화수)을 기반으로 그룹화

```python
sorted(train_test["new_num"].unique())
# 배급사별 영화수가 1개인 것부터 최대 85개 있다. 
```

```python
for i in range(86):
    a = train_test[train_test['new_num']==i].shape[0]
    if a ==0:
        pass
    else:
        print(a)

print(type(a))
# (123,50,39,44,20),(35,16,9,20,11,60,26,14,15,19,27),(36,37,38,43,76,85) -> 영화 총갯수     
# (1,2,3,4,5)       (7,8,9,10,11,12,13,14,15,19,27),  (36,37,38,43,76,85) -> 배급사별 영화갯수
# 소그룹              중그룹                  대그룹
# (276)             (252)                   (315)

# train[train['new_num']==35] 은 35개의 영화를 개봉한 배급사. 
# 데이터를 보니 5개의 배급사가 7개의 영화개봉. 5개의 배급사가 각각 7씩 총 35개의 영화
```
